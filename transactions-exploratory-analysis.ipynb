```python
import pandas as pd

successful_transactions = transactions[transactions["status"] == 0]
```

```python
successful_transactions
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'successful_transactions'
# Convert 'createdAt' to datetime
successful_transactions['createdAt'] = pd.to_datetime(successful_transactions['createdAt'])

# Extract the year and create a 'year' column
successful_transactions['year'] = successful_transactions['createdAt'].dt.year

# Create DataFrames for each year and calculate the total transactions
yearly_totals = {}  # Dictionary to store total transactions for each year
for year in range(2016, 2025):
    # Filter DataFrame for the current year
    year_df = successful_transactions[successful_transactions['year'] == year]
    
    # Calculate the total transactions for the current year
    total_transactions = year_df['amount'].sum()
    
    # Store the total transactions in the dictionary
    yearly_totals[year] = total_transactions

# Convert the yearly_totals dictionary to a DataFrame
yearly_totals_df = pd.DataFrame.from_dict(yearly_totals, orient='index', columns=['Total Transactions'])

# Create a bar plot
yearly_totals_df.plot(kind='bar', xlabel='Year', ylabel='Total Transactions', title='Total Transactions per Year')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```
